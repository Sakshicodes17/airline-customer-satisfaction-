## Label Encoding of Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
lencoders = {}
for col in data.select_dtypes(include=['object']).columns:
    lencoders[col] = LabelEncoder()
    data[col] = lencoders[col].fit_transform(data[col])
```

## Outliers Detection and Removal


```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Removing outliers from dataset
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data
```

## Feature Importance

```python
#Correlation among Features

corr = data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(150, 1, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=None, center=0,square=True, annot=True, linewidths=.5, cbar_kws={"shrink": .9})
```

```python
# Top 10 Feature Selection through Chi-Square
from sklearn import preprocessing
r_scaler = preprocessing.MinMaxScaler()
r_scaler.fit(data)

data2 = pd.DataFrame(r_scaler.transform(data), columns=data.columns)
data2.head()
```

```python
# These are top 10 features impacting on passenger satisfaction.

from sklearn.feature_selection import SelectKBest, chi2

X = data2.loc[:,data2.columns!='satisfaction']
y = data2[['satisfaction']]
selector = SelectKBest(chi2, k=10)
selector.fit(X, y)
X_new = selector.transform(X)
print(X.columns[selector.get_support(indices=True)])
```
